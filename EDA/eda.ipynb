```python
import pandas as pd
import os

# data 폴더 내의 모든 CSV 파일 목록 가져오기
data_folder = '../asesets/data'
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

# 데이터프레임을 저장할 딕셔너리
dataframes = {}

# 각 CSV 파일을 개별적으로 데이터프레임으로 불러오기
for csv_file in csv_files:
    # 파일 이름에서 확장자 제거
    df_name = os.path.splitext(csv_file)[0]
    
    # CSV 파일 경로
    file_path = os.path.join(data_folder, csv_file)
    
    # 데이터프레임 생성 및 딕셔너리에 추가
    dataframes[df_name] = pd.read_csv(file_path)
    
    print(f"'{csv_file}'을(를) '{df_name}' 데이터프레임으로 불러왔습니다.")

# 불러온 데이터프레임 확인
print("\n불러온 데이터프레임 목록:")
for df_name, df in dataframes.items():
    print(f"- {df_name}: shape {df.shape}")

# # 예시: 첫 번째 데이터프레임의 처음 5행 출력
# if dataframes:
#     first_df_name = next(iter(dataframes))
#     print(f"\n{first_df_name}의 처음 5행:")
#     print(dataframes[first_df_name].head())

# 각 데이터프레임을 전역 변수로 설정
for df_name, df in dataframes.items():
    globals()[df_name] = df
```

```python
import pandas as pd

# 오프라인 데이터 불러오기
offline_sales = pd.read_csv('./data/offline_offline_wkly_Hazzys_sales.csv')

# YYYYNN 형식의 데이터를 연도(YYYY)와 주차(NN)로 나누어 변환하는 함수
def week_to_date(week_num):
    year = int(week_num // 100)  # 앞 4자리 연도 추출
    week = int(week_num % 100)   # 뒤 2자리 주차 추출
    return pd.to_datetime(f'{year}-W{week}-1', format='%Y-W%W-%w')  # 주차의 첫번째 날로 변환

# 'Open_Store_Week0'과 'Close_Store_Week0' 변환
offline_sales['Open_Store_Week0'] = offline_sales['Open_Store_Week0'].apply(week_to_date)
offline_sales['Close_Store_Week0'] = offline_sales['Close_Store_Week0'].apply(
    lambda x: week_to_date(x) if not pd.isna(x) else pd.NaT  # NaN이면 NaT로 설정
)

# 매장별로 가장 이른 개업일(min)과 가장 늦은 폐업일(max) 선택
store_open_close = offline_sales.groupby('Store_ID').agg(
    open_date=('Open_Store_Week0', 'min'),
    close_date=('Close_Store_Week0', 'max'),
    lat=('lat', 'first'),
    lon=('lon', 'first')
).reset_index()

# 매장별 위도(lat)와 경도(lon)가 시간에 따라 변하는지 확인하는 함수
def check_location_change(store_id, df):
    store_data = df[df['Store_ID'] == store_id]
    lat_changes = store_data['lat'].nunique() > 1  # 위도 변화 여부
    lon_changes = store_data['lon'].nunique() > 1  # 경도 변화 여부
    return lat_changes or lon_changes

# 각 매장의 위치가 변동하는지 확인
store_ids_with_location_change = []

for store_id in store_open_close['Store_ID'].unique():
    if check_location_change(store_id, offline_sales):
        store_ids_with_location_change.append(store_id)

# 변동한 매장 출력
if store_ids_with_location_change:
    print(f"위치가 변동한 매장 ID: {store_ids_with_location_change}")
else:
    print("위치가 변동한 매장은 없습니다.")
```

```python
import pandas as pd
import plotly.graph_objs as go
from datetime import datetime

# 오프라인 데이터 불러오기
offline_sales = pd.read_csv('./data/offline_offline_wkly_Hazzys_sales.csv')

# YYYYNN 형식의 데이터를 연도(YYYY)와 주차(NN)로 나누어 변환하는 함수
def week_to_date(week_num):
    year = int(week_num // 100)  # 앞 4자리 연도 추출
    week = int(week_num % 100)   # 뒤 2자리 주차 추출
    return pd.to_datetime(f'{year}-W{week}-1', format='%Y-W%W-%w')  # 주차의 첫번째 날로 변환

# 'Open_Store_Week0'과 'Close_Store_Week0' 변환
offline_sales['Open_Store_Week0'] = offline_sales['Open_Store_Week0'].apply(week_to_date)
offline_sales['Close_Store_Week0'] = offline_sales['Close_Store_Week0'].apply(
    lambda x: week_to_date(x) if not pd.isna(x) else pd.NaT  # NaN이면 NaT로 설정
)

# 매장별로 가장 이른 개업일(min)과 가장 늦은 폐업일(max) 선택
store_open_close = offline_sales.groupby('Store_ID').agg(
    open_date=('Open_Store_Week0', 'min'),
    close_date=('Close_Store_Week0', 'max'),
    lat=('lat', 'first'),
    lon=('lon', 'first'),
    address=('Address', 'first'),  # 주소 추가
    store_name=('Off.Store.Name2', 'first')  # 매장 이름 추가
).reset_index()

# 명시적으로 할당하여 chained assignment 문제 해결
store_open_close['close_date'] = store_open_close['close_date'].fillna(pd.Timestamp.now())

# **월 단위**로 날짜를 변환하여 '연-월' 포맷으로 처리
store_open_close['open_month'] = store_open_close['open_date'].dt.strftime('%Y-%m')  # 개업 월
store_open_close['close_month'] = store_open_close['close_date'].dt.strftime('%Y-%m')  # 폐업 월

# 각 매장의 운영 상태에 따라 status 할당하는 함수
def assign_status(row, current_month):
    # 매장이 개업하기 전이면 '개업 전'
    if current_month < row['open_month']:
        return '개업 전'
    # 매장이 폐업한 이후에는 '폐업 전'
    elif current_month >= row['close_month']:
        return '폐업 전'
    # 매장이 운영 중일 때는 '개업 중'
    else:
        return '개업 중'

# 각 프레임별로 매장 상태(개점/폐점)를 **월 단위**로 처리하기 위해 날짜 범위 생성
min_date = store_open_close['open_date'].min()
max_date = store_open_close['close_date'].max()
month_range = pd.date_range(start=min_date, end=max_date, freq='ME')  # 월 단위로 설정 ('ME')

# 결과를 저장할 데이터프레임 생성
montly_df = pd.DataFrame()

# 한글 지도 시각화
fig = go.Figure()

# 각 월별로 매장 상태에 따라 마커 색상을 동적으로 할당
for current_date in month_range:
    current_month = current_date.strftime('%Y-%m')  # 현재 월 계산
    # 매장의 상태에 맞게 status 할당
    store_open_close['status'] = store_open_close.apply(assign_status, axis=1, current_month=current_month)
    
    # 개점 전 매장은 None으로 설정되므로, 이를 제외하고 필터링
    frame_data = store_open_close[store_open_close['status'].notna()].copy()

    # 결과 데이터프레임에 월별 상태 저장
    monthly_status = frame_data[['Store_ID', 'store_name', 'address', 'lat', 'lon', 'status']].copy()
    monthly_status['month'] = current_month
    montly_df = pd.concat([montly_df, monthly_status], ignore_index=True)

    # 운영 중 또는 폐점된 매장의 마커를 플롯에 추가 (고정된 크기로 설정)
    fig.add_trace(go.Scattermapbox(
        lat=frame_data['lat'],
        lon=frame_data['lon'],
        mode='markers',
        marker=dict(
            size=10,  # 고정된 크기
            color=frame_data['status'].map({'개업 중': 'blue', '폐업 전': 'red', '개업 전': 'white'}),
            opacity=0.8
        ),
        text=frame_data.apply(lambda row: f"매장: {row['store_name']}<br>주소: {row['address']}", axis=1),
        showlegend=False  # 날짜별로 추가된 trace는 범례에서 숨김 처리
    ))

# 한국 지도에 맞춰서 비율 조정 (세로로 길게)
fig.update_layout(
    mapbox=dict(
        center={"lat": 36.5, "lon": 127.8},  # 대한민국 중심 위치 설정
        zoom=6,  # 확대 정도 설정
        style="open-street-map"
    ),
    autosize=True,
    height=900,  # 세로로 더 길게 설정
    margin={"r": 0, "t": 0, "l": 0, "b": 0},  # 여백 제거
    showlegend=True  # 범례 표시
)

# 범례 추가 (세 가지 상태)
fig.update_traces(marker=dict(opacity=0.8), selector=dict(mode='markers'))
fig.add_trace(go.Scattermapbox(
    lat=[0], lon=[0],
    marker=dict(size=10, color='blue'),
    name="개업 중",
    showlegend=True
))
fig.add_trace(go.Scattermapbox(
    lat=[0], lon=[0],
    marker=dict(size=10, color='red'),
    name="폐업 전",
    showlegend=True
))
fig.add_trace(go.Scattermapbox(
    lat=[0], lon=[0],
    marker=dict(size=10, color='white'),
    name="개업 전",
    showlegend=True
))

# 애니메이션 재생을 위한 버튼 설정
fig.update_layout(
    updatemenus=[
        {
            "buttons": [
                {
                    "args": [None, {"frame": {"duration": 1000, "redraw": True}, "fromcurrent": True}],
                    "label": "재생",
                    "method": "animate"
                },
                {
                    "args": [[None], {"frame": {"duration": 0, "redraw": False}, "mode": "immediate", "transition": {"duration": 0}}],
                    "label": "일시정지",
                    "method": "animate"
                }
            ],
            "direction": "left",
            "pad": {"r": 10, "t": 87},
            "showactive": False,
            "type": "buttons",
            "x": 0.1,
            "xanchor": "right",
            "y": 0,
            "yanchor": "top"
        }
    ]
)

# 슬라이더로 날짜 범위를 선택할 수 있도록 설정 (월 단위 슬라이더)
fig.update_layout(sliders=[{
    "currentvalue": {"prefix": "날짜: "},
    "pad": {"b": 10},
    "steps": [{
        "args": [[str(date.strftime('%Y-%m'))]],
        "label": str(date.strftime('%Y-%m')),  # 월 단위 슬라이더
        "method": "animate"
    } for date in month_range]
}])

# Plotly fig를 HTML 파일로 저장
fig.write_html("store_animation_monthly_with_open_closed_stores_korean_map.html")

# 결과 데이터프레임을 CSV 파일로 저장
montly_df.to_csv("monthly_store_status.csv", index=False)

# 결과 데이터프레임 확인
print(montly_df.head())

# Store_ID 고유값 확인
print(f"Store ID 고유값 개수: {len(store_open_close['Store_ID'].unique())}")
```

```python
import pandas as pd
import os
from pandas_profiling import ProfileReport
import time

# data 폴더 내의 모든 CSV 파일 목록 가져오기
data_folder = './data'
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

# 데이터프레임을 저장할 딕셔너리
dataframes = {}

# 각 CSV 파일을 개별적으로 데이터프레임으로 불러오기
for csv_file in csv_files:
    # 파일 이름에서 확장자 제거
    df_name = os.path.splitext(csv_file)[0]
    
    # CSV 파일 경로
    file_path = os.path.join(data_folder, csv_file)
    
    # 데이터프레임 생성 및 딕셔너리에 추가
    dataframes[df_name] = pd.read_csv(file_path)
    
    print(f"'{csv_file}'을(를) '{df_name}' 데이터프레임으로 불러왔습니다.")

# 불러온 데이터프레임 확인
print("\n불러온 데이터프레임 목록:")
for df_name, df in dataframes.items():
    print(f"- {df_name}: shape {df.shape}")

# 각 데이터프레임을 전역 변수로 설정
for df_name, df in dataframes.items():
    globals()[df_name] = df

# EDA 보고서를 저장할 디렉토리 생성
if not os.path.exists('./eda_reports'):
    os.makedirs('./eda_reports')

# 각 데이터프레임에 대해 EDA 보고서 생성
for df_name, df in dataframes.items():
    print(f"\n{df_name}에 대한 EDA 보고서를 생성 중입니다...")
    
    # 보고서 생성
    profile = ProfileReport(df, title=f"{df_name} Profiling Report", explorative=True)
    
    # 보고서를 HTML 파일로 저장
    output_file = f"./eda_reports/{df_name}_report.html"
    profile.to_file(output_file)
    
    print(f"{df_name}의 EDA 보고서가 {output_file}에 저장되었습니다.")
    
    # 시스템 부하를 줄이기 위해 잠시 대기
    time.sleep(1)

print("\nEDA 보고서 생성이 완료되었습니다. './eda_reports' 폴더에서 결과를 확인하세요.")
```
